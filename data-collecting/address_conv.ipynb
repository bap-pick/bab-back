```python
import requests
import json
import time
```

```python
client_id = "faq1jcjgjp"
client_secret = "uvZtRjKFHCDCw9lBACgYqJy0EkTy3NPSeAV5qqHM"
```

```python
input_file = "식당.json"
output_file = "식당_좌표.json"
```

```python
with open(input_file, "r", encoding="utf-8") as f:
    restaurants = json.load(f)
```

```python
result_list = []
```

```python
for i, restaurant in enumerate(restaurants, 1):
    address = restaurant.get("address", "")
    if not address:
        continue

    endpoint = "https://maps.apigw.ntruss.com/map-geocode/v2/geocode"
    url = f"{endpoint}?query={address}"

    headers = {
        "X-NCP-APIGW-API-KEY-ID": client_id,
        "X-NCP-APIGW-API-KEY": client_secret,
    }

    res = requests.get(url, headers=headers)

    if res.status_code == 200:
        data = res.json()
        if data.get("addresses"):
            x = data["addresses"][0].get("x")  # 경도
            y = data["addresses"][0].get("y")  # 위도
            restaurant["longitude"] = x
            restaurant["latitude"] = y
        else:
            restaurant["longitude"] = None
            restaurant["latitude"] = None
    else:
        print(f"주소 변환 실패: {address}, 상태 코드: {res.status_code}")
        restaurant["longitude"] = None
        restaurant["latitude"] = None

    result_list.append(restaurant)

    # 한 식당 처리될 때마다 바로 출력
    print(f"ID: {restaurant['id']}, 주소: {restaurant['address']}, 위도: {restaurant.get('latitude')}, 경도: {restaurant.get('longitude')}")

    time.sleep(0.1)  # API 속도 제한 고려

# 전체 좌표 변환 후 JSON 파일 저장
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(result_list, f, indent=2, ensure_ascii=False)

print(f"\n총 {len(result_list)}개 식당 좌표 변환 완료. 파일 저장됨: {output_file}")
```
